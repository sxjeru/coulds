## 51.  å°†ReportCardä¸­å­¦ç”Ÿæˆç»©çš„å¹³å‡æˆç»©åˆ†ä¸ºA,B,C,D,E 5ä¸ªç­‰çº§

- A ï¼šå¤§äºç­‰äº90åˆ†ï¼›
- Bï¼šå¤§äºç­‰äº80åˆ†ï¼Œå°äº90åˆ†ï¼›
- Cï¼šå¤§äºç­‰äº70åˆ†ï¼Œå°äº80åˆ†ï¼›
- Dï¼šå¤§äºç­‰äº60åˆ†ï¼Œå°äº70åˆ†ï¼›

```r
attach(ReportCard)#è®¿é—®æ•°æ®æ¡†ä¸­åŸŸè®¿é—®
SumScore = poli + chi + math + fore + phy + che + geo + his
detach(ReportCard)
AvScore = SumScore/8#è®¡ç®—å¹³å‡å€¼
ReportCard$sumScore = SumScore
ReportCard$avScore = AvScore
# é€šè¿‡ä½¿ç”¨withinå’Œé€»è¾‘è¿ç®—ç¬¦å°†å­¦ç”Ÿå¹³å‡æˆç»©åˆ†ä¸º5ä¸ªçº§åˆ«
ReportCard = within(ReportCard,{ 
    avScore[avScore>= 90] = 'A'
    avScore[avScore>= 80 & avScore < 90] = 'B' 
    avScore[avScore>= 70 & avScore < 80] = 'C' 
    avScore[avScore>= 60 & avScore < 70] = 'D' 
    avScore[avScore < 60] = 'E' 
})
# ä¸‹é¢é€šè¿‡%in%åŒ…å«å‡½æ•°ï¼Œæ‰¾å‡ºéæ­£å¸¸é¡¹
flag = ReportCard$avScore %in% c('A',"B","C","D","E")
# é€šè¿‡ä½¿ç”¨flagæ ‡è®°ï¼Œå°†éæ­£å¸¸çš„æ ‡è®°ä¸ºç¼ºå¤±å€¼
ReportCard$avScore[!flag] = NA
# è¾“å‡ºå¹³å‡æˆç»©ç­‰çº§
print(ReportCard$avScore)
```
